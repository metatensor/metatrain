```python
# Training data
si_tiny_18 = ase.io.read("../xyz_datasets/si_tiny_ecut18.xyz", ':')
si_tiny_26 = ase.io.read("../xyz_datasets/si_tiny_ecut26.xyz", ':')
gap = ase.io.read("../../data/gap178/gap_ecut18.xyz", ':')
gap_26 = ase.io.read("../../data/gap178/gap_ecut26.xyz", ':')
si_tiny_config_types = np.loadtxt("../../data/si_tiny_types.txt", dtype=str).tolist()

# Si tiny
si_tiny_dia = []
si_tiny_bt = []
config_tipes = []
errors = []

# Choose Ecut
si_tiny = si_tiny_18
for i in range(len(si_tiny)):
    if si_tiny_config_types[i] == 'bt':
        si_tiny_bt.append(si_tiny[i])
        config_tipes.append('bt')

        error = abs(si_tiny[i].info['dft_energy'] - si_tiny_26[i].info['dft_energy'])
        errors.append(error)

    elif si_tiny_config_types[i] == 'dia':
        if len(si_tiny_dia) < 5:
            si_tiny_dia.append(si_tiny[i])
            config_tipes.append('dia')

            error = abs(si_tiny[i].info['dft_energy'] - si_tiny_26[i].info['dft_energy'])
            errors.append(error)
        else:
            continue
        
print("---------------Si_tiny---------------")
print("Beta-Sn structures: ", len(si_tiny_bt))
print("Diamond structures: ", len(si_tiny_dia))

print("-------------Additional GAP structures-------------")
gap_dia_3 = []
for at in gap:
    if at.info['config_type'] == 'dia':
        if at.info["structure_id"] in [516, 520, 536, 540, 544]: 
            gap_dia_3.append(at)
            #print(at.info["structure_id"])
            # error = abs(gap[i].info['dft_energy'] - gap_26[i].info['dft_energy'])
            # errors.append(error)
        else:
            continue 

gap26_dia_3 = []
for at in gap_26:
    if at.info['config_type'] == 'dia':
        if at.info["structure_id"] in [516, 520, 536, 540, 544]: 
            gap26_dia_3.append(at)
            #print(at.info["structure_id"])
        else:
            continue 
for i in range(len(gap_dia_3)):
    error = abs(gap_dia_3[i].info['dft_energy'] - gap26_dia_3[i].info['dft_energy'])
    print(error)
    errors.append(error)

print("GAP dia structures: ", len(gap_dia_3))
train = si_tiny_bt + si_tiny_dia #+ gap_dia_3
ref_train = si_tiny_bt + si_tiny_dia + gap26_dia_3

print("Train structures: ", len(train))
print("Errors mean: ", np.mean(errors))
print("Errors std: ", np.std(errors))
#ase.io.write("../../data/ref_train_si_tiny_bt25_dia25_gap5.xyz", ref_train)
# ase.io.write("../../data/train_si_tiny_bt25_dia25_ecut18_gap5_e18.xyz", train)
# np.savetxt("../../data/train_si_tiny_bt25_dia25_gap5_config_types.txt", config_tipes, fmt='%s')
# np.savetxt("../../data/errors/train_si_tiny_bt25_dia25_gap5_errors.txt", errors)

# GAP
gap_vacancy = []
gap_sp = []
gap_dia = []
for at in gap:
    if at.info['config_type'] == 'vacancy' :
        gap_vacancy.append(at)

    if at.info['config_type'] == 'sp':
        gap_sp.append(at)

    if at.info['config_type'] == 'dia':
        gap_dia.append(at)


ext_si_tiny = si_tiny + gap_vacancy + gap_sp + gap_dia
#ase.io.write("../../data/train_si_tiny10_gap_vac1_sp8_dia5_ecut10.xyz", ext_si_tiny)
# ase.io.write("../../data/ref_train_si_tiny26_dia3_ecut26.xyz", ext_si_tiny) #_gap_vac1_sp8_dia2_ecut26

len(si_tiny), len(gap), len(gap_vacancy), len(gap_sp), len(gap_dia), len(ext_si_tiny)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ase imports
import ase.io
```

```python
# Training data
si_tiny = ase.io.read("../xyz_datasets/si_tiny_ecut26.xyz", ':')
gap= ase.io.read("../xyz_datasets/gap_ecut26.xyz", ':')

gap_vacancy = []
for at in gap:
    if at.info['config_type'] == 'vacancy' :
        gap_vacancy.append(at)
gap_sp = []
for at in gap:
    if at.info['config_type'] == 'sp':
        gap_sp.append(at)


ext_si_tiny = si_tiny + gap_vacancy + gap_sp
#ase.io.write("../../data/ref_train_si_tiny26_gap_vac1_sp8_ecut18.xyz", ext_si_tiny)

len(si_tiny), len(gap), len(gap_vacancy), len(gap_sp), len(ext_si_tiny)
```

```python
# Train errors
gap18 = ase.io.read("../xyz_datasets/gap_ecut18.xyz", ':')
gap26 = ase.io.read("../xyz_datasets/gap_ecut26.xyz", ':')

gap18_vacancy = []
for at in gap18:
    if at.info['config_type'] == 'vacancy' :
        gap18_vacancy.append(at)
gap18_sp = []
for at in gap18:
    if at.info['config_type'] == 'sp':
        gap18_sp.append(at)
gap18_vac_sp = gap18_vacancy + gap18_sp

gap26_vacancy = []
for at in gap26:
    if at.info['config_type'] == 'vacancy' :
        gap26_vacancy.append(at)
gap26_sp = []
for at in gap26:
    if at.info['config_type'] == 'sp':
        gap26_sp.append(at)
gap26_vac_sp = gap26_vacancy + gap26_sp

errors = [0.001 for i in range(len(si_tiny))]
for i in range(len(gap18_vac_sp)):
    error = abs(gap18_vac_sp[i].info['dft_energy'] - gap26_vac_sp[i].info['dft_energy'])
    errors.append(error)
    #print(gap18[i].info['dft_energy'] - gap26[i].info['dft_energy'])

np.savetxt("../../data/gap_vac1_sp8_ecut18_errors.txt", errors)
len(errors)
```
